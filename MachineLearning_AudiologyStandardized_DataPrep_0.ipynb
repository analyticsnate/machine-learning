```python
# Import Required Packages
import pandas as pd
from pandas import Series, DataFrame

import numpy as np
```

```python
# Read col_names.txt file - this file is section 7. Attribute Information from audiology.standardized.names.txt
DF_audiology_standardized_cols = pd.read_table('col_names.txt',delimiter='\n',header=None)
S_audiology_standardized_cols = DF_audiology_standardized_cols[0]

# Column Name Cleansing
S_audiology_standardized_cols = S_audiology_standardized_cols.str.split(':').str[0] # trim rows after the colon (:)
S_audiology_standardized_cols = S_audiology_standardized_cols.str.replace('\(\)','') # remove () values
S_audiology_standardized_cols = S_audiology_standardized_cols.drop(S_audiology_standardized_cols.index[71:]) # remove rows 71+
```
