```python
import metawear_stream_threading as m
import pandas as pd
import threading
import time
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
thread1 = m.MetawearThread(10)
```

```python
thread1.start()
```

```python
plt.ion()

while thread1.isAlive():
    time.sleep(1)
    #print('running update')
    df = thread1.device_state.df.reset_index() 
    df['x'] = df['x'].astype('float64')
    #print(len(df))
    plt.plot(df['index'], df['x'])
    plt.pause(0.05)
    #print('done')
    
plt.show()
```

```python
# Create figure for plotting
plt.ion()
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

while thread1.isAlive():
    time.sleep(1)
    # get index, x values from df
    df = thread1.device_state.df.reset_index() 
    df['x'] = df['x'].astype('float64')
    # Draw index, x values
    # ax.clear()
    ax.plot(df['index'], df['x'])
    fig.canvas.draw()
```
